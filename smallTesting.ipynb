```python
import os
import soundfile as sf
import numpy as np

ROOT = "src/data/musdb18_sampled"
SUBSETS = ["train", "test"]
FILES = ["mixture.wav", "vocals.wav", "bass.wav", "drums.wav", "other.wav"]

durations = []

for subset in SUBSETS:
    subset_dir = os.path.join(ROOT, subset)
    for track in os.listdir(subset_dir):
        track_dir = os.path.join(subset_dir, track)
        if not os.path.isdir(track_dir):
            continue
        for fname in FILES:
            path = os.path.join(track_dir, fname)
            info = sf.info(path)
            durations.append(info.frames / info.samplerate)

durations = np.array(durations)

print(f"Total files       : {len(durations)}")
print(f"Min duration (s)  : {durations.min()}")
print(f"Max duration (s)  : {durations.max() }")
print(f"Mean duration (s) : {durations.mean()}")
print(f"Median duration(s): {np.median(durations)}")
print(f"Std duration (s)  : {durations.std()}")
```
